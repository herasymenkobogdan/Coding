ùëé^4 = a * a * a * a</p>

3(ùëé+ùëè)^3 + 275ùëè^2 ‚àí 127ùëé ‚àí 41</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>

<p>–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ç–æ, —á—Ç–æ –Ω–∞ —Ç–µ–∫—É—â–µ–º —ç—Ç–∞–ø–µ –æ–±—É—á–µ–Ω–∏—è –º—ã –Ω–µ –∑–Ω–∞–µ–º –ø—Ä–æ –æ–ø–µ—Ä–∞—Ç–æ—Ä –≤–æ–∑–≤–µ–¥–µ–Ω–∏—è –≤ —Å—Ç–µ–ø–µ–Ω—å,

### –°–ª–µ–¥—É—é—â–µ–µ –∏ –ø—Ä–µ–¥—ã–¥—É—â–µ–µ üî¢

–î–ª—è —á–∏—Å–ª–∞ <—Ç–µ–∫—É—â–µ–µ —á–∏—Å–ª–æ> –ø—Ä–µ–¥—ã–¥—É—â–µ–µ —á–∏—Å–ª–æ: <–ø—Ä–µ–¥—ã–¥—É—â–µ–µ —á–∏—Å–ª–æ></p>

### –°—Ç–æ–∏–º–æ—Å—Ç—å –ø–æ–∫—É–ø–∫–∏ üõí
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç–∞–µ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç—å —Ç—Ä—ë—Ö –∫–æ–º–ø—å—é—Ç–µ—Ä–æ–≤, —Å–æ—Å—Ç–æ—è—â–∏—Ö –∏–∑ –º–æ–Ω–∏—Ç–æ—Ä–∞, —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –±–ª–æ–∫–∞, –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã –∏ –º—ã—à–∏.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>

—Å—Ç–æ–∏–º–æ—Å—Ç—å –º–æ–Ω–∏—Ç–æ—Ä–∞, —Å—Ç–æ–∏–º–æ—Å—Ç—å —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –±–ª–æ–∫–∞, —Å—Ç–æ–∏–º–æ—Å—Ç—å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã, —Å—Ç–æ–∏–º–æ—Å—Ç—å –º—ã—à–∏.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>

### –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –¥–≤–∞ —á–∏—Å–ª–∞ ùëé –∏ ùëè, –≤—ã—á–∏—Å–ª—è–µ—Ç —Å—É–º–º—É, —Ä–∞–∑–Ω–æ—Å—Ç—å –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –¥–ª—è —ç—Ç–∏—Ö —á–∏—Å–µ–ª –∏ –≤—ã–≤–æ–¥–∏—Ç —Ç–µ–∫—Å—Ç –≤ —Å–ª–µ–¥—É—é—â–µ–º —Ñ–æ—Ä–º–∞—Ç–µ:<br>

<—á–∏—Å–ª–æ a> + <—á–∏—Å–ª–æ b> = <—Å—É–º–º–∞ —á–∏—Å–µ–ª a –∏ b><br>
<—á–∏—Å–ª–æ a> - <—á–∏—Å–ª–æ b> = <—Ä–∞–∑–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª a –∏ b><br>
<—á–∏—Å–ª–æ a> * <—á–∏—Å–ª–æ b> = <–ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —á–∏—Å–µ–ª a –∏ b></p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ (–∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ):
ùëé –∏ ùëè.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>

### –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—è

an = a1 + d(n ‚àí 1)</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>

## **–¶–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω–∞—è –∞—Ä–∏—Ñ–º–µ—Ç–∏–∫–∞. –ß–∞—Å—Ç—å 2**

````python
"""–ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—è
–ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–µ–π –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª ùëè1, ùëè2, ‚Ä¶, ùëèùëõ, –∫–∞–∂–¥–æ–µ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö, –Ω–∞—á–∏–Ω–∞—è —Å ùëè2,
–ø–æ–ª—É—á–∞–µ—Ç—Å—è –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —É–º–Ω–æ–∂–µ–Ω–∏–µ–º –Ω–∞ –æ–¥–Ω–æ –∏ —Ç–æ –∂–µ –ø–æ—Å—Ç–æ—è–Ω–Ω–æ–µ —á–∏—Å–ª–æ ùëû (–∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏).

–ï—Å–ª–∏ –∏–∑–≤–µ—Å—Ç–µ–Ω –ø–µ—Ä–≤—ã–π —á–ª–µ–Ω –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏ –∏ –µ–µ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å, —Ç–æ ùëõ-—ã–π —á–ª–µ–Ω –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ
ùëèùëõ = ùëè1 ‚ãÖ ùëû^(ùëõ‚àí1)

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞:
ùëè1, ùëû –∏ ùëõ , –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ùëõ-—ã–π —á–ª–µ–Ω –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏.
"""
b1 = int(input())
q = int(input())
n = int(input())
print(b1 * q ** (n - 1))
````

````python
"""–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –≤ –º–µ—Ç—Ä–∞—Ö
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –ø–æ–ª–Ω–æ–µ —á–∏—Å–ª–æ –º–µ—Ç—Ä–æ–≤ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É —á–∏—Å–ª—É —Å–∞–Ω—Ç–∏–º–µ—Ç—Ä–æ–≤.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–∞–Ω—Ç–∏–º–µ—Ç—Ä–æ–≤.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –ø–æ–ª–Ω–æ–µ —á–∏—Å–ª–æ –º–µ—Ç—Ä–æ–≤.
"""
b = int(input())
print(b // 100)
````

````python
"""–ú–∞–Ω–¥–∞—Ä–∏–Ω—ã üçä
ùëõ —à–∫–æ–ª—å–Ω–∏–∫–æ–≤ –¥–µ–ª—è—Ç k –º–∞–Ω–¥–∞—Ä–∏–Ω–æ–≤ –ø–æ—Ä–æ–≤–Ω—É, –Ω–µ–¥–µ–ª—è—â–∏–π—Å—è –æ—Å—Ç–∞—Ç–æ–∫ –æ—Å—Ç–∞–µ—Ç—Å—è –≤ –∫–æ—Ä–∑–∏–Ω–µ.
–°–∫–æ–ª—å–∫–æ —Ü–µ–ª—ã—Ö –º–∞–Ω–¥–∞—Ä–∏–Ω–æ–≤ –¥–æ—Å—Ç–∞–Ω–µ—Ç—Å—è –∫–∞–∂–¥–æ–º—É —à–∫–æ–ª—å–Ω–∏–∫—É?
–°–∫–æ–ª—å–∫–æ —Ü–µ–ª—ã—Ö –º–∞–Ω–¥–∞—Ä–∏–Ω–æ–≤ –æ—Å—Ç–∞–Ω–µ—Ç—Å—è –≤ –∫–æ—Ä–∑–∏–Ω–µ?

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∫–æ–ª—å–Ω–∏–∫–æ–≤ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∞–Ω–¥–∞—Ä–∏–Ω–æ–≤, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –¥–≤–∞ —á–∏—Å–ª–∞: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∞–Ω–¥–∞—Ä–∏–Ω–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –¥–æ—Å—Ç–∞–Ω–µ—Ç—Å—è –∫–∞–∂–¥–æ–º—É —à–∫–æ–ª—å–Ω–∏–∫—É, –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∞–Ω–¥–∞—Ä–∏–Ω–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –æ—Å—Ç–∞–Ω–µ—Ç—Å—è –≤ –∫–æ—Ä–∑–∏–Ω–µ, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.
"""
a = int(input())
b = int(input())
print(b // a)
print(b % a)
````

````python
"""–°–∞–º–∞ –Ω–µ–æ—Ç–≤—Ä–∞—Ç–∏–º–æ—Å—Ç—å
–ë–µ–∑—É–º–Ω—ã–π —Ç–∏—Ç–∞–Ω –¢–∞–Ω–æ—Å —Å–æ–±—Ä–∞–ª –≤—Å–µ 6 –∫–∞–º–Ω–µ–π –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏ –∏ –Ω–∞–º–µ—Ä–µ–≤–∞–µ—Ç—Å—è —É–Ω–∏—á—Ç–æ–∂–∏—Ç—å –ø–æ–ª–æ–≤–∏–Ω—É –Ω–∞—Å–µ–ª–µ–Ω–∏—è –í—Å–µ–ª–µ–Ω–Ω–æ–π –ø–æ —â–µ–ª—á–∫—É –ø–∞–ª—å—Ü–µ–≤.
–ü—Ä–∏ —ç—Ç–æ–º –µ—Å–ª–∏ –Ω–∞—Å–µ–ª–µ–Ω–∏–µ –í—Å–µ–ª–µ–Ω–Ω–æ–π —è–≤–ª—è–µ—Ç—Å—è –Ω–µ—á–µ—Ç–Ω—ã–º —á–∏—Å–ª–æ–º, —Ç–æ —Ç–∏—Ç–∞–Ω –ø—Ä–æ—è–≤–∏—Ç –º–∏–ª–æ—Å–µ—Ä–¥–∏–µ –∏ –æ–∫—Ä—É–≥–ª–∏—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–∂–∏–≤—à–∏—Ö –≤ –±–æ–ª—å—à—É—é —Å—Ç–æ—Ä–æ–Ω—É.
–ü–æ–º–æ–≥–∏—Ç–µ –ú—Å—Ç–∏—Ç–µ–ª—è–º –ø–æ–¥—Å—á–∏—Ç–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–∂–∏–≤—à–∏—Ö.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –¥–∞–µ—Ç—Å—è —á–∏—Å–ª–æ —Ü–µ–ª–æ–µ
n ‚Äì –Ω–∞—Å–µ–ª–µ–Ω–∏–µ –í—Å–µ–ª–µ–Ω–Ω–æ–π.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–∂–∏–≤—à–∏—Ö.
"""
a = int(input())
print(a - (a // 2))
````

````python
"""–ü–µ—Ä–µ—Å—á—ë—Ç –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ ‚åö
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –ø–µ—Ä–µ—Å—á—ë—Ç–∞ –≤–µ–ª–∏—á–∏–Ω—ã –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞, –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –≤ –º–∏–Ω—É—Ç–∞—Ö, –≤ –≤–µ–ª–∏—á–∏–Ω—É, –≤—ã—Ä–∞–∂–µ–Ω–Ω—É—é –≤ —á–∞—Å–∞—Ö –∏ –º–∏–Ω—É—Ç–∞—Ö –≤ —Å–ª–µ–¥—É—é—â–µ–º —Ñ–æ—Ä–º–∞—Ç–µ:

<–∏—Å—Ö–æ–¥–Ω–æ–µ –∫–æ–ª-–≤–æ –º–∏–Ω—É—Ç> –º–∏–Ω - —ç—Ç–æ <–ø–æ–ª–Ω–æ–µ –∫–æ–ª-–≤–æ —á–∞—Å–æ–≤> —á–∞—Å <–æ—Å—Ç–∞–≤—à–µ–µ—Å—è –∫–æ–ª-–≤–æ –º–∏–Ω—É—Ç> –º–∏–Ω—É—Ç.
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.
"""
a = int(input())
print(f"{a} –º–∏–Ω - —ç—Ç–æ {a // 60} —á–∞—Å {a % 60} –º–∏–Ω—É—Ç.")
````

````python
"""–¢—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç —Å—É–º–º—É –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–≥–æ —á–∏—Å–ª–∞ –∏ –≤—ã–≤–æ–¥–∏—Ç —Ç–µ–∫—Å—Ç –≤ —Å–ª–µ–¥—É—é—â–µ–º —Ñ–æ—Ä–º–∞—Ç–µ:

–°—É–º–º–∞ —Ü–∏—Ñ—Ä = <—Å—É–º–º–∞ —Ü–∏—Ñ—Ä>
–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä = <–ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä>
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.
"""
a = int(input())
last = a % 10
mid = (a // 10) % 10
fir = a // 100
print("–°—É–º–º–∞ —Ü–∏—Ñ—Ä =", last + mid + fir)
print("–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä =", last * mid * fir)
````

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–µ —Å—Ç—Ä–æ–∫–∏.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.</p>

### –ß–µ—Ç–Ω–æ–µ –∏–ª–∏ –Ω–µ—á–µ—Ç–Ω–æ–µ?
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è —á–∏—Å–ª–æ —á–µ—Ç–Ω—ã–º –∏–ª–∏ –Ω–µ—á–µ—Ç–Ω—ã–º.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´–ß–µ—Ç–Ω–æ–µ¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫), –µ—Å–ª–∏ —á–∏—Å–ª–æ —á–µ—Ç–Ω–æ–µ, –∏–ª–∏ ¬´–ù–µ—á–µ—Ç–Ω–æ–µ¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫) –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.</p>

````python
a = int(input())
if a % 2 == 0:
    print("–ß–µ—Ç–Ω–æ–µ")
else:
    print("–ù–µ—á–µ—Ç–Ω–æ–µ")
````

### –°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —á—Ç–æ –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–µ—Ç—ã—Ä–µ—Ö–∑–Ω–∞—á–Ω–æ–≥–æ —á–∏—Å–ª–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–µ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ: —Å—É–º–º–∞ –ø–µ—Ä–≤–æ–π –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–π —Ü–∏—Ñ—Ä —Ä–∞–≤–Ω–∞ —Ä–∞–∑–Ω–æ—Å—Ç–∏ –≤—Ç–æ—Ä–æ–π –∏ —Ç—Ä–µ—Ç—å–µ–π —Ü–∏—Ñ—Ä.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–µ—Ç—ã—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´–î–ê¬ª, –µ—Å–ª–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –∏ ¬´–ù–ï–¢¬ª ‚Äî –µ—Å–ª–∏ –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.</p>

````python
a = int(input())
last = a % 10
mid2 = (a // 10) % 10
mid1 = (a // 100) % 10
fir = a // 1000
if fir + last == mid1 - mid2:
    print("–î–ê")
else:
    print("–ù–ï–¢")
````

### –†–æ—Å–∫–æ–º–Ω–∞–¥–∑–æ—Ä üîû
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —Ä–∞–∑—Ä–µ—à—ë–Ω –ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –¥–æ—Å—Ç—É–ø –∫ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ä–µ—Å—É—Ä—Å—É –∏–ª–∏ –Ω–µ—Ç.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äì –≤–æ–∑—Ä–∞—Å—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç ¬´–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω¬ª, –µ—Å–ª–∏ –≤–æ–∑—Ä–∞—Å—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –Ω–µ –º–µ–Ω–µ–µ 18 –ª–µ—Ç, –∏–ª–∏ ¬´–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.</p>

````python
a = int(input())
if a < 18:
    print("–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω")
else:
    print("–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω")
````

### –ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—è
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —è–≤–ª—è—é—Ç—Å—è –ª–∏ —Ç—Ä–∏ –∑–∞–¥–∞–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ (–≤ —É–∫–∞–∑–∞–Ω–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ) –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã–º–∏ —á–ª–µ–Ω–∞–º–∏ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —Ç—Ä–∏ —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´YES¬ª –∏–ª–∏ ¬´NO¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫) –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.</p>

````python
a, b, c = int(input()), int(input()), int(input())
if (a + c) / 2 * 3 != a + b + c:
    print("NO")
else:
    print("YES")
````

### –ù–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ –¥–≤—É—Ö —á–∏—Å–µ–ª
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ –¥–≤—É—Ö —á–∏—Å–µ–ª.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ –¥–≤—É—Ö —á–∏—Å–µ–ª.</p>

````python
a, b = int(input()), int(input())
if a < b:
    print(a)
else:
    print(b)
````

### –ù–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ —á–µ—Ç—ã—Ä—ë—Ö —á–∏—Å–µ–ª üå∂Ô∏è
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ —á–µ—Ç—ã—Ä—ë—Ö —á–∏—Å–µ–ª.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —á–µ—Ç—ã—Ä–µ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ —á–µ—Ç—ã—Ä—ë—Ö —á–∏—Å–µ–ª.</p>

<p>–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –£—á–∏—Ç—ã–≤–∞–π—Ç–µ, —á—Ç–æ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ –º–æ–≥—É—Ç –¥—É–±–ª–∏—Ä–æ–≤–∞—Ç—å—Å—è (—Å–º–æ—Ç—Ä–∏—Ç–µ —Ç–µ—Å—Ç ‚Ññ5). –í —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–≤–µ—Å—Ç–∏ —Ç–æ–ª—å–∫–æ –æ–¥–Ω–æ –∏–∑ –Ω–∏—Ö.</p>

````python
a, b, c, d = int(input()), int(input()), int(input()), int(input())
# print(min(a, b, c, d))

if a < b:
    b = a
else:
    a = b

if c < d:
    d = c
else:
    c = d

if a < d:
    d = a
else:
    a = d
print(a)
````

### –í–æ–∑—Ä–∞—Å—Ç–Ω–∞—è –≥—Ä—É–ø–ø–∞ üëßüèª
<p>–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–º—É –≤–æ–∑—Ä–∞—Å—Ç—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —Å–æ–æ–±—â–∞–µ—Ç, –∫ –∫–∞–∫–æ–π –≤–æ–∑—Ä–∞—Å—Ç–Ω–æ–π –≥—Ä—É–ø–ø–µ –æ–Ω –æ—Ç–Ω–æ—Å–∏—Ç—Å—è:<br>
–¥–æ 13 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ) ‚Äì –¥–µ—Ç—Å—Ç–≤–æ;<br>
–æ—Ç 14 –¥–æ 24 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ) ‚Äì –º–æ–ª–æ–¥–æ—Å—Ç—å;<br>
–æ—Ç 25 –¥–æ 59 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ) ‚Äì –∑—Ä–µ–ª–æ—Å—Ç—å;<br>
–æ—Ç 60 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ) ‚Äì —Å—Ç–∞—Ä–æ—Å—Ç—å.</p>

<p>–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>

<p>–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö<br>

````python
a = int(input())
if a <= 13:
    print("–¥–µ—Ç—Å—Ç–≤–æ")
elif 13 < a <= 24:
    print("–º–æ–ª–æ–¥–æ—Å—Ç—å")
elif 24 < a <= 59:
    print("–∑—Ä–µ–ª–æ—Å—Ç—å")
else:
    print("—Å—Ç–∞—Ä–æ—Å—Ç—å")

# print("—Å—Ç–∞—Ä–æ—Å—Ç—å" if n>59 else "–∑—Ä–µ–ª–æ—Å—Ç—å" if n>24 else "–º–æ–ª–æ–¥–æ—Å—Ç—å" if n>13 else "–¥–µ—Ç—Å—Ç–≤–æ")
````
